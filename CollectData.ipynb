```python
#csv.register_dialect('unixpwd', delimiter=':', quoting=csv.QUOTE_NONE)
def read1DFile(filePath):
    with open(filePath, mode='r',) as csv_file:
        csv_reader = csv.reader(csv_file)
        nTimePoints = sum(1 for row in csv_reader) 
        csv_file.seek(0)
        line_count = 0
        data = np.zeros((nTimePoints,90))
        for row in csv_reader:
            if line_count == 0:
                #print(f'Column names are {"- ".join(row)}')
                line_count += 1

            else:
                tokens = nltk.word_tokenize(row[0])
                #print("Processing line {}".format(line_count-1))
                for j in range(0,90):
                    data[line_count-1][j] = float(tokens[j])
                    #print ("{}-".format(j), end='')
                    
                #print("line count is {}".format(line_count))
                line_count += 1
    return np.transpose(data)
```

```python
def countTimePoints(filePath):
    with open(filePath, mode='r',) as csv_file:
        csv_reader = csv.reader(csv_file)
        nTimePoints = sum(1 for row in csv_reader) 
        csv_file.seek(0)
        
    return nTimePoints
```
